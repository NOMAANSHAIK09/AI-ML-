```python
import json
def load_data(filen):
    with open(filen,"r") as f:
        data = json.load(f)
    return data
```

```python
data=load_data("store_data.json")
print(data)
```

```python
#cleaning ratedata
def cleaning_data(data):
    text_to_num = {"one":1,"two":2,"three":3,"four":4,"five":5}
    clean_data=[]
    unique_data = set()

    for user in data:
        raw_rating = str(user["rating"]).strip().lower()
        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
        user["rating"]=raw_rating

        raw_age = user.get("age")
        
        if (raw_age == None):
            user["age"]= None

        if(user["name"].strip() in unique_data):
            continue
        unique_data.add(user["name"])
        clean_data.append(user)
        fresh_data = clean_data
        
    return fresh_data



cleaning_data(data)

```
