```python
import json
def load_data(file):
    with open(file,"r") as f:
        data = json.load(f)
    return data

df = load_data("freaf_data.json")
print(df)
```

```python
#avg rating and % user poor rating
def get_insight(df):
    tot_rating = 0
    for user in df:
        tot_rating += float(user["rating"])
    print(f"avg rating={tot_rating/len(df)}")

    poor_rating=0
    for user in df:
        if(float(user["rating"]) < 3):
            poor_rating +=1
    print(f"poor rating = {(poor_rating/len(df))*100}%")
            

get_insight(df)
```
