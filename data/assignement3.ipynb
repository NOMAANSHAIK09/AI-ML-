```python

import json
def load_data(file):
    with open(file,"r") as f:
        data = json.load(f)
    return data

df = load_data("freaf_data.json")
print(df)
```

```python
#user >= 4 and <4 recommandation brand4
def get_recomm(df):
    recomm=[]
    for user in df:
        cur_recom={}
        cur_recom["name"]=user["name"]
        if(float(user["rating"]) >= 4):
            cur_recom["brand"]="apple"
        else:
            cur_recom["brand"]="samsung"

        recomm.append(cur_recom)
    return recomm

get_recomm(df)
```
